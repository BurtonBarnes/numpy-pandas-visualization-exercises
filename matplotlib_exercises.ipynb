```python
#3.
n_rows = 2
n_cols = 2

x = [1,2,3,4,5,6]
y = [x**(1/2) for x in x]
z = [(x*x*x) for x in x]
w = [(2**x) for x in x]
v = [1/(x+1) for x in x]

plt.subplot(n_rows, n_cols, 1)
plt.plot(x, y)
plt.title('x ~ y')

plt.subplot(n_rows, n_cols, 2)
plt.plot(x, z)
plt.title('x ~ z')

plt.subplot(n_rows, n_cols, 3)
plt.plot(x, w)
plt.title('x ~ w')

plt.subplot(n_rows, n_cols, 4)
plt.plot(x, v)
plt.title('x ~ v')
```

```python
#4.
plt.figure(figsize = (8,2))
n_rows = 1
n_cols = 2

x = [1,2,3,4,5,6]
y = [x**(1/2) for x in x]
z = [(x*x*x) for x in x]
w = [(2**x) for x in x]
v = [1/(x+1) for x in x]

plt.subplot(n_rows, n_cols, 1)
plt.plot(x, y, label='x^(1/2)')
plt.subplot(n_rows, n_cols, 1)
plt.plot(x, z, label='x^3')
plt.subplot(n_rows, n_cols, 1)
plt.plot(x, w, label='2^x')
plt.subplot(n_rows, n_cols, 1)
plt.plot(x, v, label='1/(x+1)')
plt.title('Multiple plots in one figure')

plt.legend(loc='upper right')
plt.ylim(0,5)
plt.show()
```
